```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from pandas.tseries.offsets import MonthEnd
import warnings
warnings.filterwarnings('ignore')      #Отключим предупреждения об устаревающих методах
pd.options.display.max_rows = 500      #Увеличим кол-во выводимых строк, чтобы просматривать сводные таблицы целиком
```

```python
df_courses = pd.read_csv('courses.csv', index_col='Unnamed: 0')
df_students = pd.read_csv('students.csv')
df_course_contents = pd.read_csv('course_contents.csv')
df_progresses = pd.read_csv('progresses.csv')
df_progress_phases = pd.read_csv('progress_phases.csv')
```

```python
main_1 = df_progress_phases.merge(df_progresses, how='left', left_on='progress_id', right_on='id')
main_1.drop(['id'], axis=1, inplace=True)
main_2 = main_1.merge(df_courses, how='left', left_on='course_id', right_on='id')
main_2.drop(['id'], axis=1, inplace=True)
main = main_2.merge(df_course_contents, how='left', on=['course_id', 'module_number', 'lesson_number'])
main['start_date'] = pd.to_datetime(main['start_date']).dt.date
main['finish_date'] = pd.to_datetime(main['finish_date']).dt.date   
```

```python
main.head(3)
```

*Посчитаем количество курсов*

```python
main['title'].nunique()
```

*Посчитаем количество модулей (modules_quantity) и количество студентов (students_quantity) на каждом курсе*

```python
courses_info = main.groupby(['title']).agg(modules_quantity=('module_title','nunique'),
                                           students_quantity=('student_id','nunique'))
courses_info
```

*Посчитаем кол-во уроков в каждом курсе на каждом модуле*

```python
lesson_quantity = main.groupby(['title', 'module_title']).agg(lesson_title=('lesson_title','nunique'))      
lesson_quantity
```

*Посчитаем медианное количество уроков в модуле на каждом курсе (lesson_count_median). Дополним сводную таблицу*

```python
courses_info['lesson_count_median'] = lesson_quantity.groupby('title').median()
courses_info
```

*Определим максимальный, минимальный, средний и медианный возраст студентов*

*Сначала создадим датафрейм для работы с датой рождения. В столбце "city" много пропущенных значений. Удалим его, он пока не нужен. Затем удалим строки, в которых дата рождения не заполнена.*

```python
df_students_age = df_students.drop(['city'], axis=1)
df_students_age.dropna(inplace=True)
```

*Удалим из него ошибочные даты и приведем к формату datetime*

```python
for i, row in df_students_age.iterrows():
    year = row['birthday'][:4]
    year = int(year)
    if year < 1970 or year > 2005:
        df_students_age.drop([i], axis=0, inplace=True)

df_students_age['birthday'] = pd.to_datetime(df_students_age['birthday'])
```

*Добавим столбец с возрастом*

```python
def calc_age(birthday):
    today = dt.date.today()
    return today.year - birthday.year - ((today.month, today.day) < (birthday.month, birthday.day))

df_students_age['age'] = df_students_age['birthday'].apply(calc_age)
```

```python
df_students_age.head(2)
```

*Узнаем возраст студентов*

```python
print('Минимальный возраст студетов {} лет'.format(df_students_age['age'].min()))
print('Максимальный возраст студентов {} лет'.format(df_students_age['age'].max()))
print('Средний возраст студентов {} лет'.format(round(df_students_age['age'].mean(), 1)))
print('Медианное значение возраста студентов {} лет'.format(df_students_age['age'].median()))
```

*Посчитаем минимальный (age_min), максимальный (age_max), средний (age_mean), медианный (age_median) возраст студентов по каждому курсуу*

```python
students_info = main.drop_duplicates(subset=['student_id'])
students_info = students_info.merge(df_students_age, how='inner', left_on='student_id', right_on='id')
students_info = students_info.groupby(['title']).agg(age_min=('age', 'min'), age_max=('age', 'max'),
                                                        age_mean=('age', 'mean'), age_median=('age', 'median'))
    
courses_info = courses_info.merge(students_info, on='title')
courses_info['field'] = main.groupby('title')['field'].first()
courses_info
```

*Построим график отражающий кол-во студентов на каждом курсе. На графике укажем медиану количества студентов.*

```python
new = courses_info.groupby(['title', 'field'])['students_quantity'].first().unstack().fillna(0)
new['sort'] = new['Business']+new['Development']+new['Marketing']+new['Design']
new = new.sort_values('sort', ascending=True).drop(['sort'], axis=1)
fig = plt.figure(figsize=(10,6))
plt.box(on=None)
for name in new.columns:
    plt.barh(new.index, new[name], label=name, alpha=0.3)
plt.title('Количество студентов на каждом курсе', size=18)
plt.axvline(courses_info['students_quantity'].median(), color='black')
plt.legend()
plt.show()
```

*Узнаем первый модуль с домашней работой для каждого курса*

```python
first_mod = df_course_contents[df_course_contents['is_homework'] == True].groupby('course_id')['module_number'].min().reset_index()
first_mod = first_mod.merge(df_courses[['title', 'id']], left_on='course_id', right_on='id')[['title', 'module_number']]
first_mod
```

*Рассчитаем прирост студентов на каждом курсе. Выберем из основного датасета только записи с выполнением домашней работы и узнаем месяц начала выполнения первой домашней работы у каждого студента (month_tag)*

```python
data = main[main['is_homework'] == True]
#сoздадим столбец указывающий на конец месяца.
data['month_tag'] = data['start_date'] + MonthEnd(0)
#Узнаем время начала каждого модуля
data = data.groupby(['title', 'student_id', 'module_number'])['month_tag'].min().reset_index()
#Оставим только модули с домашней работой
data = data.merge(first_mod, on=['title', 'module_number'])
data
```

*Посчитаем кол-во новых студетнов в каждый месяц*

```python
courses_growth = data.groupby('title')['month_tag'].value_counts().unstack().T.reset_index().fillna(0)
courses_growth
```

*Построим график отражающий количество новых студентов каждый месяц по каждому курсу*

```python
courses_growth['month_tag'] = courses_growth['month_tag'].dt.date
fig,ax = plt.subplots(figsize=[15,7])
plt.box(on=None)
ax.set_title('Прирост студентов по каждому курсу', size= 17)
ax.set_ylabel('Прирост студентов', size=12)
for name in courses_growth.columns[1:]:
    ax.plot(courses_growth.index, courses_growth[name], label=name)
ax.set_xticks(courses_growth.index)
ax.set_xticklabels(courses_growth['month_tag'], rotation=45)
ax.legend()
plt.show()
```

*Посчитаем кол-во прогрессов в каждом месяце для каждого курса*

```python
#Нам нужны только записи с домашней работой.
data = main[main['is_homework'] == True][['start_date', 'finish_date', 'title', 'status', 'progress_id']]
#Создадим столбец указывающий на конец месяца по аналогии с предыдущим заданием.
data['month_tag'] = data['start_date'] + MonthEnd(0)
#Будем считать, что не выполненые задания были проверены один раз в месяц начала выполнения. Создадим сводную таблицу для них.
grow = data[data['status'] != 'done'].groupby(['title', 'month_tag'])['progress_id'].count().unstack().T.fillna(0)
#Затем удалим эти записи из датасета. 
data = data[data['status'] == 'done']
#Далее узнаем кол-во прогрессов которые не вышли за рамки одного месяца
grow_1 = data[(data['start_date'] + MonthEnd(0)) == (data['finish_date'] + MonthEnd(0))]\
                        .groupby(['title', 'month_tag'])['progress_id'].count().unstack().T.fillna(0)
#Потом удалим их
data = data[(data['start_date'] + MonthEnd(0)) != (data['finish_date'] + MonthEnd(0))]
#И посчитаем прогрессы которые охватили 2 или больше месяцев
def f(x):  #Используя функцию создадим список по месяцам
    z = list(pd.date_range(x['start_date'], x['finish_date'], freq='M').date)
    a = x['finish_date'] + MonthEnd(0)
    a = pd.to_datetime(a).date()
    z.append(a)
    return z

data['month_tag'] = data.apply(f, axis=1)
data = data.explode('month_tag')
grow_2 = data.groupby(['title', 'month_tag'])['start_date'].count().unstack().T.fillna(0)

#Осталось сложить три датасета
grows = grow.add(grow_1, fill_value=0).add(grow_2, fill_value=0).reset_index()
grows
```

*По данным из полученной таблицы построим график отражающий кол-во прогрессов для каждого курса по месяцам*

```python
grows['month_tag'] = grows['month_tag'].dt.date
fig,ax = plt.subplots(figsize=[15,7])
plt.box(on=None)
plt.title('Количество прогрессов по каждому курсу', size= 17)
plt.ylabel('Прирост', size=12)
for name in grows.columns[1:]:
    ax.plot(grows.index, grows[name], label=name)
ax.set_xticks(grows.index)
ax.set_xticklabels(grows['month_tag'], rotation=45)
plt.legend()
plt.show()
```

*Основная масса курсов, например по Java-Script, начали набирать популярность в феврале 2018г. *Вычислим время прохождения каждого модуля*

## 4. Расчет конверсии

1. **(p)** Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).

3. **(m)** Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. 15 графиков. Прозрачность должна стоять на отметке 0.1. На графики должна быть нанесена линия медианы конверсии для каждого курса. У медианы должен быть свой цвет. Рамок у графиков быть не должно ⭐
4. На основании рассчитанных значений сделайте аналитический вывод (должен быть полным и покрывать все полученные выше метрики)

*У нас уже есть таблица первого и последего урока с ДЗ в каждом модуле. Добавим в нее название каждого модуля*

```python
lessons = df_course_contents[df_course_contents['is_homework'] == True].groupby(['course_id', 'module_number', 'module_title'])\
                                .agg(lesson_max=('lesson_number','max'), lesson_min=('lesson_number','min')).reset_index()
lessons = lessons.merge(df_courses[['title', 'id']], left_on='course_id', right_on='id')\
                                [['title', 'module_number',  'module_title','lesson_max', 'lesson_min']]
lessons
```

```python
#Посчитаем кол-во студентов прошедших последний урок в каждом модуле.
data = main[(main['is_homework'] == True)&(main['status'] == 'done')] 
data = data.groupby(['title', 'module_number', 'lesson_number']).agg(students_finish=('student_id','count')).reset_index()
#Узнаем количество студентов начавших первый урок в модуле
data_2 = main[main['is_homework'] == True] 
data_2 = data_2.groupby(['title', 'module_number', 'lesson_number']).agg(students_start=('student_id','count')).reset_index()
```

```python
#Объединим все данные в один датафрейм и вычислим конверсию
convers = data.merge(lessons, how='inner', left_on=['title','module_number', 'lesson_number'],
                      right_on=['title','module_number', 'lesson_max'])\
                        [['title','module_number', 'module_title', 'lesson_number','students_finish']]

convers2 = data_2.merge(lessons, left_on=['title','module_number', 'lesson_number'],
                    right_on=['title','module_number', 'lesson_min'])\
                                [['title','module_number', 'module_title', 'lesson_number','students_start']]

convers = convers.merge(convers2, on=['title','module_number', 'module_title'])\
                        [['title','module_number', 'module_title', 'students_finish','students_start']]
convers_res = convers.copy()
convers_res['students_finish'] = convers_res['students_finish'].shift(+1)
convers_res['convers'] = convers_res['students_start'] / convers_res['students_finish']

convers_res

```

*Построим столбчатую диаграмму отражающую конверсию по каждому курсу*

```python
fig = plt.figure(figsize=(20,300))
n = 1
for name, values in convers.groupby('title'):
    values['module_number'] = values['module_number'].astype('str')
    values['names'] = values['module_title'] + '  №' + values['module_number']
    values['students_finish'] = values['students_finish'].shift(+1)
    values['convers'] = values['students_start'] / values['students_finish']
    ax = fig.add_subplot(15, 1, n)
    ax.barh(values['module_title'], values['convers'], alpha=0.2, color='g')
    ax.set_title(name, size=35)
    plt.box(on=None)
    ax.axvline(values['convers'].median())
    ax.set_yticklabels(values['names'], size=25)
    plt.xticks(size=25)
    n += 1
    
plt.show()


    
```

*Конверсия для многих курсов колеблется, в том числе поднимается выше единицы. Думаю это связано с тем, что все модули изначально открыты для прохождения и у студентов есть возможность оставлять их на потом и проходить в случайном порядке*
